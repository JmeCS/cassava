```python
df = pd.read_csv(path/'train.csv')
df.head()
```

```python
df.label.unique()
```

```python
cassava = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    splitter=RandomSplitter(seed=42),
    get_x=ColReader(0, pref=path/"train_images"),
    get_y=ColReader(1),
    item_tfms=Resize(460), #not sure what is best here
    batch_tfms=aug_transforms(size=244, min_scale=0.57)
)
```

```python
dls = cassava.dataloaders(df)
```

```python
dls.show_batch()
```

```python
cassava.summary(path/'train_images')
```

```python
learn = cnn_learner(dls, resnet34, metrics=error_rate)
```

```python
learn.fine_tune(3)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix(figsize=(12,12))
```

```python
interp.most_confused(min_val=50)
```

```python
# Need to reset learner before using lr_find()
learn = cnn_learner(dls, resnet34, metrics=error_rate)
lr_min, lr_steep = learn.lr_find()
```

```python
print(f"Minimum/10: {lr_min:.2e}, steepest point: {lr_steep:.2e}")
```

```python
learn = cnn_learner(dls, resnet34, metrics=error_rate)
learn.fine_tune(2, base_lr=8.32e-3)
```

```python
# Train last layer(s) only for 3 cycles
learn = cnn_learner(dls, resnet34, metrics=error_rate)
learn.fit_one_cycle(3, 8.32e-3)
```

```python
# unfreeze all layers and find new learning rate 
learn.unfreeze()
learn.lr_find()
```

```python
# Use discriminative LR 
learn.fit_one_cycle(12, lr_max=slice(1.9e-6, 1e-5))
```

```python
learn.fit_one_cycle(8, lr_max=slice(1.9e-6, 1e-5))
```

```python
learn.fit_one_cycle(4, lr_max=slice(1.9e-6, 1e-5))
```

```python
learn.save('res34')
```
